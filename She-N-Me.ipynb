```python
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'Sachin' and 'Media' not in word
                                and not word.startswith('media')
                                and not word.startswith('omitted')
                                and not word.startswith('None')
                                and not word.startswith('Will')
                                and not word.startswith('Know')
                            ])
    
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()

train_pos = new['Text']
wordcloud_draw(train_pos.astype('str'),'black')
```
